```python
board = [list(input()) for _ in range(9)]
def cross(A, B):
    return [a+b for a in A for b in B]

digits = '123456789'
rows = 'ABCDEFGHI'
cols = digits
squares=cross(rows,cols)
# print(s)
unitlist = ([cross(rows, c) for c in cols] + [cross(r, cols) for r in rows] + [cross(rs, cs) for rs in ('ABC', 'DEF', 'GHI') for cs in ('123', '456', '789')])
# print(unitlist)
units = dict((s, [u for u in unitlist if s in u]) for s in squares)
# print(units)
peers = dict((s, set(sum(units[s],[]))-set([s])) for s in squares)
# print(peers)

def grid_values(board):
    chars = []
    for l in board:
        temp = [c for c in l if c in digits or c in '0.']
        chars += temp
    return dict(zip(squares, chars))

def parse_grid(board): ## Values(좌표:digits)
    values = dict((s, digits) for s in squares)
    print(values)
    for s, d in grid_values(board).items():  ##grid_values(board) => 스도쿠 숫자 맵핑
        if d in digits and not assign(values, s, d): #d라는 숫자가 들어 갔을때 같은 값이 있는지 판정 (and 앞에 False 면 더이상 진행 안됨)
            return False
    print(values)
    return values

def assign(values, s, d): 
    other_values = values[s].replace(d, '') 
    if all(eliminate(values, s, d2) for d2 in other_values):  #같은 값 없는 경우에 return values 
        return values
    else:
        return False

def eliminate(values, s, d): #s좌표가 아닌 d값
    if d not in values[s]:
        return values
    values[s] = values[s].replace(d, '')
    if len(values[s]) == 0:
        return False
    elif len(values[s]) == 1:
        d2 = values[s]
        if not all(eliminate(values, s2, d2) for s2 in peers[s]):
            return False
    for u in units[s]:
        dplaces = [s for s in u if d in values[s]]
        if len(dplaces) == 0:
            return False
        elif len(dplaces) == 1:
            if not assign(values, dplaces[0], d):
                return False
    return values

def display(values):
    width = 1+max(len(values[s]) for s in squares)
    for r in rows:
        print(' '.join(values[r+c] for c in cols))

def solve(board): return search(parse_grid(board)) ## 시작

def search(values):
    if values is False:
        return False
    if all(len(values[s]) == 1 for s in squares):
        return values
    n,s = min((len(values[s]), s) for s in squares if len(values[s]) > 1)
    return some(search(assign(values.copy(), s, d)) for d in values[s]) 

def some(seq):
    for e in seq:
        if e: return e
    return False

display(solve(board))
```
